```python
# import
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) 

#Read in the data

df = pd.read_csv(r'C:\Users\dania\OneDrive\Рабочий стол\Study\movies.csv')
```

```python
df.head()
```

```python
# LEt's if there is any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df.dtypes
```

```python
df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
print("Пропуски в budget:", df['budget'].isna().sum())
print("Пропуски в gross:", df['gross'].isna().sum())
```

```python
total_rows = len(df)
print("budget: {} NaN ({}%)".format(
    df['budget'].isna().sum(),
    round(df['budget'].isna().mean()*100, 2)
))
print("gross: {} NaN ({}%)".format(
    df['gross'].isna().sum(),
    round(df['gross'].isna().mean()*100, 2)
))

```

```python
df
```

```python
df=df.sort_values(by=['gross'], inplace=False, ascending=False)
df
```

```python
#pd.set_option('display.max_rows', None)
pd.reset_option('display.max_rows')
```

```python
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#Scatter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
#Plot budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df,scatter_kws={'color': 'red'}, line_kws={'color': 'blue'} )
plt.show()

```

```python
# Correlation
df.corr(method='pearson', numeric_only=True) #pearson, kendall, spearman
```

```python
correlation_matrix = df.corr(method='pearson', numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
df_cat = df

for col_cat in df_cat.columns:
    if (df_cat[col_cat].dtype == ('object')):
        df_cat[col_cat] = df_cat[col_cat].astype('category')
        df_cat[col_cat] = df_cat[col_cat].cat.codes
df_cat
```

```python
correlation_matrix = df_cat.corr(method='pearson', numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
cormatrix = df_cat.corr(method='pearson', numeric_only=True)
corr_pairs = cormatrix.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs 
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
